```python
import json
import pandas as pd
from pandas import json_normalize

def read_and_concat_json(files):
    """
    複数のJSONファイルを読み込んで行方向に連結されたDataFrameを返す関数

    Parameters:
    - files: list, 読み込むJSONファイルのリスト

    Returns:
    - DataFrame: 行方向に連結されたDataFrame
    """
    dfs = []  # DataFrameを格納するリスト

    for file in files:
        with open(file, 'r') as f:
            data = json.load(f)

        # 必要な列のみを選択してDataFrameに変換
        target_columns = ['name', 'Accuracy', 'Precision', 'Recall', 'F1-score', 'Auc', 'Rules.violation', 'Rules.total']
        df = pd.json_normalize(data)[target_columns]

        # "Violation ratio" カラムを追加
        df['Violation ratio (= violation / total)'] = df['Rules.violation'] / df['Rules.total']

        # DataFrameをリストに追加
        dfs.append(df)

    # リスト内のDataFrameを行方向に連結
    result_df = pd.concat(dfs, ignore_index=True)

    return result_df

# 複数のJSONファイルがある場合、ファイル名をリストとして指定
json_files = ['result_4.json', 'result_3.json', 'result_5.json', 'result_6.json', 'result_0.json', 'result_1.json', 'result_2.json', 'result_7.json', 'result_8.json']

# 関数を呼び出してDataFrameを取得
combined_df = read_and_concat_json(json_files)


print('-')

combined_df
```
